```python
from ultralytics import YOLO

# Load a YOLO11n PyTorch model
model = YOLO("yolo11n.pt")

# Export the model to TensorRT
model.export(format="engine")  # creates 'yolo11n.engine'

# Load the exported TensorRT model
trt_model = YOLO("yolo11n.engine")

# Run inference
results = trt_model("https://ultralytics.com/images/bus.jpg")
```

```python
import cv2
from ultralytics import YOLO
from ultralytics.utils.plotting import Annotator


# model = YOLO("yolo11n.pt")

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    results = trt_model(frame)

    annotator = Annotator(frame, line_width=1, example=str(results[0].names))

    for det in results[0]:
        x_center, y_center, w, h = det.boxes.xywh.cpu().numpy()[0]
        confidence = det.boxes.conf.cpu().numpy()[0]
        class_idx = det.boxes.cls.cpu().numpy()[0]
        
        
        if confidence > 0.5:
            x1, y1 = int((x_center - w / 2)), int((y_center - h / 2))
            x2, y2 = int((x_center + w / 2)), int((y_center + h / 2))
            annotator.box_label((x1, y1, x2, y2), label=det.names[int(class_idx)])

    cv2.imshow('0', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
